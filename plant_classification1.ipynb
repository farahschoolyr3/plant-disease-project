```python
import matplotlib.pylab as plt
import tensorflow as tf
import tensorflow_hub as hub

import os
import numpy as np

import tensorflow_datasets as tfds
import cv2
import warnings
warnings.filterwarnings('ignore')
```

```python
#Total Count of Images
data_path=os.listdir()
vimages=0
traimages=0
cate=['Pepper__bell___Bacterial_spot',
 'Pepper__bell___healthy',
 'Potato___healthy',
 'Potato___Late_blight',
 'Tomato_Bacterial_spot',
 'Tomato_healthy',
 'Tomato_Late_blight',
 'Tomato__Tomato_mosaic_virus']
for d in data_path:
    if d=='validation':
        for c in cate:
            imgs=os.listdir(d+'/'+c)
            vimages+=len(imgs)
    elif d=='train':
        for c in cate:
            imgs=os.listdir(d+'/'+c)
            traimages+=len(imgs)
```

```python
vimages-=8
traimages-=8
```

```python
print(vimages)
print(traimages)
```

```python
#Labels Array intiallizing with 1 for every Image
train_labels=np.ones((traimages,),dtype='int64')
validation_labels=np.ones((vimages,),dtype='int64')
```

```python
#labelling Images
val_list=[]
test_list=[]
train_list=[]
h=w=224
for d in data_path:
    if d=='validation':
        v=0
        a=0
        for c in cate:
            imgs=os.listdir(d+'/'+c)
            for i in imgs:
                try:
                    val_data=cv2.imread(d+'/'+c+'/'+i)
                    print(d+'/'+c+'/'+i)
                    #val_data=cv2.cvtColor(val_data,cv2.COLOR_BGR2GRAY)
                    #h, w = val_data.shape
                    val_data=cv2.resize(val_data,(h,w))
                    print('validation  '+str(v)+'  '+str(val_data.shape))
                    val_list.append(val_data)

                    if c==cate[0]:
                        validation_labels[v]=0
                    elif c==cate[1]:
                        validation_labels[v]=1
                    elif c==cate[2]:
                        validation_labels[v]=2
                    elif c==cate[3]:
                        validation_labels[v]=3
                    elif c==cate[4]:
                        validation_labels[v]=4
                    elif c==cate[5]:
                        validation_labels[v]=5
                    elif c==cate[6]:
                        validation_labels[v]=6
                    elif c==cate[7]:
                        validation_labels[v]=7
                    v+=1
                except:
                    print('It was not an Image')
    elif d=='train':
        v=0
        a=0
        for c in cate:
            imgs=os.listdir(d+'/'+c)
            for i in imgs:
                try:
                    train_data=cv2.imread(d+'/'+c+'/'+i)
                    print(d+'/'+c+'/'+i)
                    #train_data=cv2.cvtColor(train_data,cv2.COLOR_BGR2GRAY)
                    #h, w = train_data.shape
                    train_data=cv2.resize(train_data,(h,w))
                    print('train  '+str(v)+'  '+str(train_data.shape))
                    train_list.append(train_data)

                    if c==cate[0]:
                        train_labels[v]=0
                    elif c==cate[1]:
                        train_labels[v]=1
                    elif c==cate[2]:
                        train_labels[v]=2
                    elif c==cate[3]:
                        train_labels[v]=3
                    elif c==cate[4]:
                        train_labels[v]=4
                    elif c==cate[5]:
                        train_labels[v]=5
                    elif c==cate[6]:
                        train_labels[v]=6
                    elif c==cate[7]:
                        train_labels[v]=7

                    v+=1
                except:
                    print('It was not an Image')
            

```

```python
train_images=np.array(train_list)
train_images=train_images.astype('float32')
train_images/=255
print(train_images.shape)
#train_images=np.expand_dims(train_images,axis=3)
print(train_images.shape)

validation_images=np.array(val_list)
validation_images=validation_images.astype('float32')
validation_images/=255
#validation_images=np.expand_dims(validation_images,axis=3)
print(validation_images.shape)
```

```python

from keras.utils import np_utils
train_label=np_utils.to_categorical(train_labels,8)
validation_label=np_utils.to_categorical(validation_labels,8)
```

```python
feature_extractor = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"
```

```python
feature_extractor_layer = hub.KerasLayer(feature_extractor, input_shape=(224,224,3))
```

```python
feature_extractor_layer.trainable = False
```

```python
model = tf.keras.Sequential([
  feature_extractor_layer,
  tf.keras.layers.Dropout(0.3),
  tf.keras.layers.Dense(8,activation='softmax')
])

model.summary()
```

```python
model.compile(
  optimizer=tf.keras.optimizers.Adam(),
  loss=tf.keras.losses.CategoricalCrossentropy(from_logits=True),
  metrics=['acc'])
```

```python
history = model.fit(train_images,train_label, epochs=30, validation_data=(validation_images,validation_label))                        
```

```python
result=model.evaluate(validation_images,validation_label)
```

```python
for i in range(10)  :
    image, label = validation_images[i], validation_label[i]
    #image_scaled, label_arr= scale(validation_images, test_sample[1])
    image_scaled = np.expand_dims(image, axis=0)   
    print(image_scaled.shape)
    #img = tf.keras.preprocessing.image.img_to_array(image_scaled)                    
    pred=model.predict(image_scaled)
    print(pred)
    plt.figure()
    plt.imshow(image)
    plt.show()
    print("Actual Label: %s" % validation_labels[i])
    print("Predicted Label: %s" % np.argmax(pred))
```
